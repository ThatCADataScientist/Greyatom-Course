```
a = np.zeros([4, 4])
b = np.ones([4, 4])
print(a)
print(b)
```

```
# to concatenate rows
dvg = np.concatenate([a, b], axis=0)  
podu = np.vstack([a,b])  
didi = np.r_[a,b]
print(dvg) 
print(podu)
print(didi)

# to concatenate columns
dvg = np.concatenate([a, b], axis=1)  
podu = np.hstack([a,b])  
didi = np.c_[a,b]
print(dvg) 
print(podu)
print(didi)
```

```
x = np.array([1, 10, 5, 2, 8, 9])
sort_index = np.argsort(x)
print(sort_index)
#How to interpret this?
#In array ‘x’, the 0th item is the smallest, 3rd item is the second smallest and so on.
```

```
# Get the index positions that would sort the array
x = np.array([1, 10, 5, 2, 8, 9])
sort_index = np.argsort(x)
sort_index1 = np.argsort(-x)
print(sort_index)
print(sort_index1)
x[sort_index]
```

```
x = np.array([1,1,2,2,2,4,4,5,6,6,6])
np.bincount(x)
```

```
# Histogram example - in this counts is used to count the number of digits in the bin ...also bin is a range eg 0-2, 2-4,4-6,6-8    
counts, bins = np.histogram(x, [0, 2, 4, 6, 8])
print('Counts: ', counts)
print('Bins: ', bins)
```
