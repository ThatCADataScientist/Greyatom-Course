```
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 1.0 + 0.01, 0.01)
s = np.cos(2 * 2*np.pi * t)
t[41:60] = np.nan

plt.subplot(1, 2, 1)
plt.plot(t, s, '-', lw=2)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('A sine wave with a gap of NaNs between 0.4 and 0.6')
plt.grid(True)

plt.subplot(1, 2, 2)
t[0] = np.nan
t[-1] = np.nan
plt.plot(t, s, '-', lw=2)
plt.title('Also with NaN in first and last point')

plt.xlabel('time (s)')
plt.ylabel('more nans')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```
#SUBPLOTS
#to show more than one graph together
fig,axs = plt.subplots(2,3)
axs[0,0].plot(worldh["Rank"]**2,worldh["Score"],"green")
axs[0,1].plot(worldh["Rank"],worldh["Score"]**3,"Red")
axs[0,2].plot(worldh["Rank"]*5,worldh["Score"],"Orange")
axs[1,1].plot(worldh["Rank"],worldh["Score"]+5)#axs(1,0)&(2,0) are kept empty to show differences
```

```
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 1.0 + 0.01, 0.01)
s = np.cos(2 * 2*np.pi * t)
t[41:60] = np.nan

for i in range(1,10):
  plt.subplot(3,3,i)
  plt.plot(t, s, '-', lw=2)
  

plt.show()           
```
